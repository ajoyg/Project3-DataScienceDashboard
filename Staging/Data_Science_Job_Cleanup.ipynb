```python
import pandas as pd
import numpy as np
import geopandas as gpd
from geopy.geocoders import Nominatim
```

```python
ds_jobs_raw_df = pd.read_csv('../Resources/Cleaned_DS_Jobs.csv')
ds_jobs_raw_df
```

```python
geolocator = Nominatim(timeout=5, user_agent="geocoder")  
ds_jobs_raw_df['geocode'] = ds_jobs_raw_df.loc[:,"Location"].apply(geolocator.geocode)
ds_jobs_raw_df
```

```python
ds_jobs_raw_df["latitude"] = [g.latitude for g in ds_jobs_raw_df.geocode]
ds_jobs_raw_df["longitude"] = [g.longitude for g in ds_jobs_raw_df.geocode]
ds_jobs_raw_df.head()
```

```python
ds_jobs_raw_df['Sector'] = ds_jobs_raw_df['Sector'].str.replace('- ', '_')
ds_jobs_raw_df['Sector'].head(50)
```

```python
values_to_drop=['United States', 'Patuxent, Anne Arundel, MD']
ds_jobs_all_city = ds_jobs_raw_df.loc[~ds_jobs_raw_df['Location'].isin(values_to_drop)]
ds_jobs_all_city.head(50)
ds_jobs_all_city[['city', 'state']] = ds_jobs_all_city['Location'].str.split(',', expand=True)
ds_jobs_all_city
```

```python
ds_jobs_all_city.info()
```

```python
df_jobs_cleaned = ds_jobs_all_city.drop(["geocode","Headquarters"], axis=1)
df_jobs_cleaned = df_jobs_cleaned.rename(columns= {"Job Title": "job_title","Salary Estimate": "salary_estimate","Job Description": "job_description","Rating": "rating",\
                                                   "Company Name": "company_name","Location": "location", "Headquarters": "headquarters", "Size": "size","Type of ownership": "type_of_ownership",\
                                                    "Industry": "industry", "Sector": "sector", "Revenue": "revenue"})
df_jobs_cleaned = df_jobs_cleaned[['job_title','salary_estimate','rating','company_name','city','state','latitude','longitude',\
                                   'size','type_of_ownership','industry','sector','revenue','min_salary','max_salary','avg_salary','job_state','same_state','company_age',\
                                    'python','excel','hadoop','spark','aws','tableau','big_data','job_simp','seniority']]

df_jobs_cleaned
```

```python
#Get the Salaries data
df_salaries_raw = pd.read_csv("../Resources/ds_salaries.csv")
df_salaries_raw
```

```python
df_salaries_cleaned = df_salaries_raw[['work_year', 'experience_level','employment_type','job_title','salary','salary_currency', 'salary_in_usd','employee_residence','remote_ratio','company_location','company_size']]
df_salaries_cleaned
```

```python
df_salaries_cleaned['company_location'].unique()
df_salaries_raw['geocode'] = df_salaries_raw.loc[:,"company_location"].apply(geolocator.geocode)
df_salaries_raw
```

```python
df_salaries_raw["latitude"] = [g.latitude for g in df_salaries_raw.geocode]
df_salaries_raw["longitude"] = [g.longitude for g in df_salaries_raw.geocode]
df_salaries_raw
```

```python
df_salaries_cleaned = df_salaries_raw.drop(columns=['Unnamed: 0', 'geocode'])
df_salaries_cleaned = df_salaries_cleaned[['work_year', 'experience_level','employment_type','job_title','salary','salary_currency', 'salary_in_usd','employee_residence','remote_ratio','company_location','latitude','longitude','company_size']]
df_salaries_cleaned
```

```python
# Save the cleaned files a a csv into the Staging folder
df_jobs_cleaned.to_csv("data/jobs_cleaned.csv", index=False)
df_salaries_cleaned.to_csv("data/salaries_cleaned.csv", index=False)
```
